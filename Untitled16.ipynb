```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

corr = df[[
    "Sleep Duration",
    "Quality of Sleep",
    "Stress Level",
    "Physical Activity Level",
    "Heart Rate",
    "Daily Steps",
    "Age"
]].corr()

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)

plt.title("Overall Correlation Between Sleep and Lifestyle Factors")

plt.show()
```

```python
fig, ax = plt.subplots(2,2, figsize=(14,10))

df.groupby("Occupation")["Sleep Duration"].mean().plot(
    kind="bar", ax=ax[0,0],
    title="Sleep Duration"
)

df.groupby("Occupation")["Quality of Sleep"].mean().plot(
    kind="bar", ax=ax[0,1],
    title="Quality of Sleep"
)

df.groupby("Occupation")["Stress Level"].mean().plot(
    kind="bar", ax=ax[1,0],
    title="Stress Level"
)

df.groupby("Occupation")["Physical Activity Level"].mean().plot(
    kind="bar", ax=ax[1,1],
    title="Physical Activity"
)

plt.tight_layout()
plt.show()
```
